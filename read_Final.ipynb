```python
import pandas as pd
```

```python
# Read in the first CSV:
df1 = pd.read_csv("fake_data_address.csv")
df1.head(5)
```

```python
df2 = pd.read_csv("fake_data_dob.csv")
df2.head(5)
```

```python
# Create a 'full_name' column that we can use to merge the two dataframes:
df1 = df1.assign(full_name = df1.last_name + ", " + df1.first_name + " " + df1.middle_name) 
df1.head(5)
```

```python
# Rename unweildy column name to 'full_name':
df2['full_name'] = df2['last_name, first_name middle_name']
df2.head(5)
```

```python
# Merge the two dataframes on the 'full_name' column:
df3 = pd.merge(df1, df2, on='full_name')
df3.head(5)
```

```python
# Create a new 'line1' address column for line 1 of the address:
re =  r'^Line 1: (.*)(?:Line 2: )'
df3['line1'] = df3['Line 1: address Line 2: address2'].str.extract(re)
df3.head(5)
```

```python
# Create a new 'line2' column for the second line of the address:
re =  r'.*Line 2: (.*)'
df3['line2'] = df3['Line 1: address Line 2: address2'].str.extract(re)
df3.head(5)
```

```python
# Create a new 'city' column for the city:
re = r'-(.*)'
df3['city'] = df3['state-city'].str.extract(re)
df3.head(5)
```

```python
# Create a new 'state' column for the state:
re = r'(.*)-'
df3['state'] = df3['state-city'].str.extract(re)
df3.head(5)
```

```python
# Convert the zip to an int, so it prints better:
df3['zip'] = df3['zip'].astype(int)
df3.head(5)
```

```python
# Convert the date of birth column to a datetime object:i
df3['dob'] = pd.to_datetime(df3['dob'])
df3.head(5)
```

```python
# Validate the data types for each column:
df3.info()
```

```python
# Convert the date specified by Eli to a datetime object:
from datetime import datetime
date_string = "3/21/2023"
date_format = "%m/%d/%Y"
date_obj = datetime.strptime(date_string, date_format)
print(date_obj)
```

```python
# Add an age column by subtracting our date from the date of birth:
# 
df3['age'] = (date_obj - df3['dob']).astype('<m8[Y]')
df3.head(5)
```

```python
# Clean up the dataframe by dropping unneeded columns:
# 
# There are two ways to drop columns: 
# df3.drop(['last_name,  firstname middle_name'], axis=1)
# df3.drop(columns=['last_name, first_name middle_name'])
df4 = df3.drop(columns=['Line 1: address Line 2: address2', 'state-city', 'full_name', 'last_name, first_name middle_name'])
df4.info()
```

```python
# Create new dataframe using ages between 40 and 60, inclusive:
age_df = df4.loc[(df4["age"] >= 40) & (df4["age"] <= 60) , :]
age_df.info()
```
